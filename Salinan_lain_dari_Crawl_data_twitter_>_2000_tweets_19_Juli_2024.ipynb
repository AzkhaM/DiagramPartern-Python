Merge

```python
import os
import pandas as pd

# Path folder tempat file CSV disimpan
folder_path = 'merge-data'

# List semua file CSV dalam folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Gabungkan semua file CSV
merged_df = pd.concat([pd.read_csv(os.path.join(folder_path, file)) for file in csv_files], ignore_index=True)

# Simpan hasil gabungan ke file baru
merged_df.to_csv('merged_tweets_data.csv', index=False)

print("Penggabungan selesai! File disimpan sebagai 'merged_tweets_data.csv'")
```

Ngilangin grok

```python
#KODE UNTUK ILANGIN LINK DI FULL TEXT + ILANGIN GROK + HAPUS DUPLIKAT

import re
import pandas as pd
from google.colab import files

import pandas as pd

# Ganti nama file sesuai dengan file aslinya
filename = 'merged_tweets_data.csv'

# Baca file CSV
df = pd.read_csv(filename)

# Jumlah sebelum deduplikasi
print("\nJumlah baris sebelum menghapus duplikat+GROK:", len(df))

# Hapus baris yang berisi kata "saya grok" di kolom full_text
df = df[~df['full_text'].astype(str).str.contains('Saya Grok', na=False)]

# Simpan versi tanpa "saya grok"
output_filename = 'cleaned_no_saya_grok_' + filename
df.to_csv(output_filename, index=False)

# Hapus tautan dari kolom full_text
df['full_text'] = df['full_text'].apply(lambda x: re.sub(r'https?://\S+', '', x) if pd.notnull(x) else x)

# Simpan ke file baru
output_filename = 'cleaned_' + filename
df.to_csv(output_filename, index=False)  # Removed sep='\t'


# Cek data duplikat berdasarkan kolom 'full_text'
duplicate_rows = df[df.duplicated(subset=['full_text'], keep=False)]

# Tampilkan data duplikat
print("Data duplikat berdasarkan 'full_text':")
display(duplicate_rows)



# Hapus data duplikat
df_cleaned = df.drop_duplicates(subset=['full_text'], keep='first')

# Jumlah sesudah deduplikasi
print("Jumlah baris setelah menghapus duplikat:", len(df_cleaned))

# Simpan hasil akhir
df_cleaned.to_csv('HASIL AKHIR1.csv', index=False)
```
